```python
# Treasure Game

# Give a hint if the treasure is behind one of the doors
def give_hint(doors):
    if 'Treasure' in doors:
        return "You sense something valuable behind one of these doors!"
    return "You feel danger lurking nearby" 

def treasure_game():
    print('Welcome to a Treasure game\nThis game consists of two levels "Easy and Hard"\nYou can choose one of them\nEnjoy while Playing')

    # Asking the user to choose a difficulty level
    while True:
        check_game_level = input('Choose the level of the game "Easy" or "Hard": ').lower()
        if check_game_level == 'easy':
            traits = 10  # Easy level gives 10 attempts
            break
        elif check_game_level == 'hard':
            traits = 5  # Hard level gives 5 attempts
            break
        else:
            print('Invalid choice. Please select either "Easy" or "Hard".')

    score = 0
    level = 1        

    print(f'You have {traits} attempts to find the treasure.')    

    # Game loop for choosing a door
    while traits > 0:
        print(f'The Level is {level}')

        # Fixed outcomes behind the doors for each level
        doors = ['Lion', 'Treasure', 'empty']  # This sequence remains constant

        # Option to ask for a hint (costs 2 attempts)
        if input('Do you want a hint? (yes/no): ').lower() == 'yes':
            print(give_hint(doors))
            traits -= 2  # Decrease attempts when using a hint
            print(f'You have {traits} attempts left')

        choice_doors = input('Please choose from these doors: "Door1", "Door2", or "Door3": ').lower()
        # Check the outcome based on fixed doors
        if doors[0] == 'Lion' and choice_doors == 'door1':
            print('You encountered a lion and lost the game')
            return
        elif doors[1] == 'Treasure' and choice_doors == 'door2':
            print('Congratulations, you found the treasure and won this level!')
            score += 100
            print(f'Your score is now: {score}')
            traits += 3  # Add extra attempts as a reward for winning a level
            level += 1   # Progress to the next level
        elif doors[2] == 'empty' and choice_doors == 'door3':
            print('This door is empty. Keep trying!')
        else:
            print('Invalid choice. Please select "Door1", "Door2", or "Door3"')

        traits -=1
        print(f'You have {traits} attempts left.')  

    # If the user runs out of attempts
    print('You have no more attempts. Game over!') 
    print(f'Your final score is: {score}')

treasure_game()
```
